```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Pictures/MSD_8193.jpg',1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img,cmap='gray')
```

```python
img1 = cv2.Sobel(img,cv2.CV_64F,1,0,1) #horizontal sobel
img2 = cv2.Sobel(img,cv2.CV_64F,0,1,1) #vertical sobel

plt.imshow(img1,cmap='gray'),plt.title('Vertical Sobel')
```

```python
plt.imshow(img1,cmap='gray'),plt.title('Horizontal Sobel')
```

```python
img3 = cv2.Laplacian(img,cv2.CV_64F)
plt.imshow(img3,cmap='gray'),plt.title('Laplcaian')
```

```python
img4 = cv2.Canny(img,100,250)
plt.imshow(img4,cmap='gray'),plt.title('Canny')
```
